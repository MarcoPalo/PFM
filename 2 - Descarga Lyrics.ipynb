En este notebook descargamos las letras de las canciones. 

Usamos Google porque tiene varias bases de datos, y porque aún con erratas (al tener símbolos y caracteres de distintos idiomas es difícil hacer un match exacto) encuentra el resultado.

Google tiene automatizado el salto de una ventana emergente cada vez que entras para que no puedas hacer requests, por lo que la forma de llegar a la información es automatizando el proceso, como con los rankings de Spotify. Además, el mensaje que sale varía entre 2, por lo que en la automatización hay metido un reconocimiento de imagen para actuar en consecuencia. 

Este método descarga el .html de la búsqueda de Google y lo guarda individualmente. En el notebook anterior, está el código de cómo se saca la letra desde ahí.

La última celda solo tiene el output de las primeras 10 líneas ya que he vueltoa a correrlo para confirmar que sigue funcionando. El proceso para realizar las 49544 búsquedas tarda varios días y tiene que hacerse en primera plana.

```python
songs = pd.read_excel('total_songs.xlsx')
songs
```

```python
#Descarga Lyrics Google
for i in range (0,10):
    text = str(songs.loc[i,'Song'])+' '+str(songs.loc[i,'Artist'])+' '+ ' lyrics'
    print(str(i) + '    ' + text)

    driver = webdriver.Chrome() #abrimos Chrome
    driver.get('https://www.google.com') #vamos a Google
    time.sleep(1)

    #Aceptamos el consentimiento
    try: ## miramos si es el tipo de mensaje de nuestra captura y aceptamos
        x , y = pyautogui.locateCenterOnScreen("bien2.jpg", confidence = 0.95)

        pyautogui.hotkey('ctrl', 'l')
        for i in range (0,6):
            pyautogui.hotkey('tab')
        pyautogui.hotkey('enter')
        time.sleep(0.3)

    except: ##llegamos al aceptar para el caso en el que no es nuestra captura 
        pyautogui.hotkey('ctrl', 'l') 
        for i in range (0,8):
            pyautogui.hotkey('tab')
        pyautogui.hotkey('enter')
        time.sleep(0.3)

    pyautogui.hotkey('ctrl', 'l') #Buscamos los lyrics de la canción
    for i in range (0,10):
        pyautogui.hotkey('tab')
    pyautogui.typewrite(text)
    pyautogui.hotkey('enter')
    time.sleep(1)  

    pyautogui.hotkey('ctrl', 's') #Descargamos el html
    time.sleep(0.5)
    pyautogui.hotkey('tab')
    pyautogui.press('right')
    pyautogui.press('up')
    pyautogui.press('up')
    pyautogui.hotkey('enter')
    pyautogui.hotkey('enter')
    time.sleep(1.5)

    pyautogui.hotkey('alt','f4') #cerramos ventana (no es lo más rápido del mundo tener que cerrar ventana a cada iteración
    ##pero es la manera de que Google no nos bloquee como bot)
```
