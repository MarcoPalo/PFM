Este código sirve para descargar los rankings de Spotify. Requiere usuario y contraseña. No existe una forma de hacerlo mediante requests, por lo que utilizamos técnicas de automatización con selenium y pyautogui. El proceso es el siguiente: abrimos Chrome (requiere tener el último webdriver en el directorio) con Selenium. Nos metemos en la página de Spotify Charts. Con pyautogui vamos automatizando el proceso de llegar hasta el inicio de sesión, de meter nuestros datos, iniciar sesión y aceptar las cookies. Después nos vamos metiendo en la url de cada ranking con Selenium. Para cada url, automatizamos con pyautogui la descarga (hay un botón con el que descargamos un csv con los datos del ranking). El proceso contiene tiempos de espera para el cargado de las páginas, las cuales son ligeramente aleatorias para dificultar que se nos detecte como un bot.

Si se quiere comprobar el funcionamiento, se debe meter un user (nombre del correo) , mail(dominio del correo) y password (contraseña de Spotify). Comentar que no todas las páginas de los charts son iguales, por lo que a veces hay que rehacer el código cambiando el número de veces a las que hay que dar al shift en el último loop para poder dar con el botón de descargar.
