### Top 10

```python
df = pd.read_csv('total2.csv')
df = df[df['Position']<10]
df.reset_index(inplace=True, drop=True)
df
```

```python
#Streams por semana y país (para calcular el coeficiente ponderado)
streams_per_week = df.groupby(['date','region']).agg({'Streams':['sum']})
streams_per_week.rename(columns = {'sum':'','Streams':'streams_region10'},inplace=True)
streams_per_week.reset_index(inplace=True)
streams_per_week['date']=streams_per_week['date'].astype(str)
streams_per_week['id']=streams_per_week['region']+'_'+streams_per_week['date']
streams_per_week.drop({'region','date'}, axis=1, inplace=True)
streams_per_week
```

```python
df['date2']=df['date'].astype(str)
df['id']=df['region']+'_'+df['date2']
df = pd.merge(df, streams_per_week, on='id', how='left') 
df.rename({('streams_region10', ''):'streams_region10'},axis=1,inplace=True)
df.drop({'date2'},axis=1,inplace=True)
df
```

```python
df['top10']=(df['positive2']*df['Streams'])/df['streams_region10']
```

```python
for i in range(len(df)):
    df.loc[i,'month']=df.loc[i,'date'][0:7]+'-01'
df
```

```python
top10_per_month = df.groupby(['month','region']).agg({'top10':['mean']})
top10_per_month.rename(columns = {'mean':'','Streams':'top10'},inplace=True)
top10_per_month.reset_index(inplace=True)
top10_per_month['month']=top10_per_month['month'].astype(str)
top10_per_month['id']=top10_per_month['region']+'_'+top10_per_month['month']
top10_per_month['top10']=((top10_per_month['top10'] - top10_per_month['top10'].min()) / 
                         (top10_per_month['top10'].max() - top10_per_month['top10'].min()))*100 
top10_per_month.drop({'region','month'}, axis=1, inplace=True)
top10_per_month
```

```python
panel = pd.read_excel('panel_final2.xlsx')
panel = pd.merge(panel, top10_per_month, on='id', how='left') 
panel.rename({('top10', ''):'top10'},axis=1,inplace=True)
panel
```

```python
panel.to_excel('panel_final2.xlsx',index=False)
```

### Top100

```python
df = pd.read_csv('total2.csv')
df = df[df['Position']<100]
df.reset_index(inplace=True, drop=True)
df
```

```python
#Streams por semana y país (para calcular el coeficiente ponderado)
streams_per_week = df.groupby(['date','region']).agg({'Streams':['sum']})
streams_per_week.rename(columns = {'sum':'','Streams':'streams_region100'},inplace=True)
streams_per_week.reset_index(inplace=True)
for i in range(len(streams_per_week)):
    streams_per_week.loc[i,'date']=str(streams_per_week.loc[i,'date'])[4:14]
streams_per_week['id']=streams_per_week['region']+'_'+streams_per_week['date']
streams_per_week.drop({'region','date'}, axis=1, inplace=True)
streams_per_week
```

```python
for i in range(len(df)):
    df.loc[i,'date2']=str(df.loc[i,'date'])[0:10]
df['id']=df['region']+'_'+df['date2']
df = pd.merge(df, streams_per_week, on='id', how='left') 
df.rename({('streams_region100', ''):'streams_region100'},axis=1,inplace=True)
df.drop({'date2'},axis=1,inplace=True)
df['top100']=(df['positive2']*df['Streams'])/df['streams_region100']
for i in range(len(df)):
    df.loc[i,'month']=df.loc[i,'date'][0:7]+'-01'
df
```

```python
top100_per_month = df.groupby(['month','region']).agg({'top100':['mean']})
top100_per_month.rename(columns = {'mean':'','Streams':'top100'},inplace=True)
top100_per_month.reset_index(inplace=True)
top100_per_month['month']=top100_per_month['month'].astype(str)
top100_per_month['id']=top100_per_month['region']+'_'+top100_per_month['month']
top100_per_month['top100']=((top100_per_month['top100'] - top100_per_month['top100'].min()) / 
                         (top100_per_month['top100'].max() - top100_per_month['top100'].min()))*100 
top100_per_month.drop({'region','month'}, axis=1, inplace=True)
top100_per_month
```

```python
panel = pd.read_excel('panel_final2.xlsx')
panel = pd.merge(panel, top100_per_month, on='id', how='left') 
panel.rename({('top100', ''):'top100'},axis=1,inplace=True)
panel
```

```python
panel.to_excel('panel_final2.xlsx',index=False)
```
